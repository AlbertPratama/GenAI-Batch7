## `Basic Time Series Manipulation`

```python
from datetime import datetime

date_rng = pd.date_range(start='1/01/2020', end='1/08/2020', freq='H')
date_rng
```

```python
df = pd.DataFrame(date_rng, columns=['date'])
df['data'] = np.random.randint(low=0, high=100, size=(len(date_rng)))
df_time_series = df
```

```python
df_time_series.info()
```

```python
df_time_series.set_index('date', inplace=True)
df_time_series
```

```python
# Menggunakan .query()
result_query = df_time_series.query("date == '2020-01-08 00:00:00'")
result_query
```
