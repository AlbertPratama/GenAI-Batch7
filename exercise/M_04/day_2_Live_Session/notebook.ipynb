```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
alice_novel = open('../../datasets/alice_novel.txt', 'r').read()
```

```python
stopwords = set(STOPWORDS)
stopwords.add('said')
```

```python
alice_wc = WordCloud(background_color='white', max_words=200, stopwords=stopwords)
alice_wc.generate(alice_novel)
```

```python
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
stop_word = set(STOPWORDS)
stopwords.add('went')
```

```python
alice_wc = WordCloud(background_color='black',
                     max_words=200, stopwords=stopwords)
alice_wc.generate(alice_novel)

plt.figure(figsize=[12, 12], facecolor='k')
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

# `Choropleth Map`

```python
df_can = pd.read_csv('Canada.csv')
# df_can.set_index('Country', inplace=True)
df_can.head()
```

```python
world_geo = r'../../datasets/world_countries.json'
```

```python
world_map = folium.Map(
    location=[0, 0],
    zoom_start=2,
)

folium.Choropleth(
    geo_data=world_geo,
    data=df_can,
    columns=['Country', 'Total'],
    key_on='feature.properties.name',
    fill_color='YlOrRd',  # Yellow, Orange, Red
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Immigration to Canada'
).add_to(world_map)

world_map
```
