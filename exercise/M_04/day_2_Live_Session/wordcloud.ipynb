# Wordcloud

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
import matplotlib.pyplot as plt
```

```python
alice_novel = open('alice_novel.txt', 'r').read()
```

```python
# setup wordcloud
alice_wc = WordCloud(background_color='white',
                     max_words=2000,
                     stopwords=STOPWORDS)

alice_wc.generate(alice_novel)
```

```python
plt.imshow(alice_wc)
```

```python
stopwords = set(STOPWORDS)
stopwords.add('said')
```

```python
# setup wordcloud
alice_wc = WordCloud(background_color='black',
                     max_words=2000,
                     stopwords=stopwords,
                     colormap='Blues'
                     )

alice_wc.generate(alice_novel)


plt.figure(figsize=(12, 12), facecolor='k')
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from PIL import Image
import numpy as np
```

```python
alice_mask = np.array(Image.open('alice_mask.png'))
```

```python
# show masking

plt.figure(figsize=(14, 18))
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# setup wordcloud
alice_wc = WordCloud(background_color='black',
                     max_words=2000,
                     stopwords=stopwords,
                     colormap='Blues',
                     mask=alice_mask
                     )

alice_wc.generate(alice_novel)


plt.figure(figsize=(12, 12), facecolor='k')
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
